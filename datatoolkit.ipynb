1. What is the difference between multithreading and multiprocessing?
 - 1. Memory:- threads share memory ; processes have separate memory spaces .
   2. isolated & safety :- process are isolated(safer,less accidental data corruption); however threads needs locks to avoid race conditions.
   3. communication :- threads communicate via shared variables(with locks); processes use IPC(queues,pipes,shared memory).
   4. Crash impact :- if a thead crashes it can make the whole process down; if a process crashes other usually continue.

2. What are the challenges associated with memory management in Python?
-  1. Reference cycles - objects referring to each other prevent automatic cleanup - possible memory leaks.
 2. lingering reference - unused objects kept in lists, golbals, or caches are not memory freed.
 3. memory fragmentation - frequent allocation/ deallocation causes scattered memory blocks- wasted space.
 4. high object overhead - Each python object uses extra memory for metadata(inefficient forlarge data)
 5. GIL & Threads - shared memory acroos threads can cause race condition or inconsistent states.
 6. Delayed destruction - __del__() may not run immediately- resources (files,sockets) stay open.
 7. Memory not returned to OS - freed memory may stay reserved by python,s allocated.
 8. External leaks - c extension or libraries may allocate memory outside python's control.

6.  What are the differences between NumPy arrays and Python lists?
 - 1. data type -  numPy arrays store only one data type, lsits can hold different types.
  2. speed - numpy arrays are faster for mathematical operations.
  3. memory - numpy arrays use less memory than lists.
  4. operations - arrays support vectorized(element-wise) operation; lists need loops.
  5. Dimenions- arrays can be multi- dimensional; lists are mainly one- dimensional.
  6. functions - numpy provides many built-in mathematical function; lists do not.
  7. use case - use lists for general data storage and arrays for scientific or numerical computing.

7. Explain the difference between apply() and map() in Pandas.
  - 1. Purpose = **map()** is used only with series(a single column).
**apply()** can be used with both series and dataframes(entire rows or column).
  2. scope =  **map()** works element-wise on a series. it applies a function, dictionary, or mapping to each value.
**apply()** works on rowss or column of a dataframe - it can apply a custom function to each row or column.
  3.return type = **map()** always returns a new series.
  **apply()** can return a series, dataframes, or scalar, depennding on the function.

```python
#Question 10: Calculate the correlation matrix using Seaborn and visualize it with a
 # heatmap
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Maths': [78, 85, 96, 80, 86],
    'Science': [84, 94, 89, 83, 86],
    'English': [70, 80, 75, 88, 90]
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("correlation heatmap of subjects")
plt.show()
```
